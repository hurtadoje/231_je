__Genome annotation and Pseudomonas writeup__

Pseudomonas is a genus of gram-negative, rod-shaped, motile, and aerobic bacteria that have been isolated from animals, soils, and plants from around the world  [1].  Its ability to survive in animals including humans, as well as its antibiotic resistance as evidenced by the presence of the arpC and arpB antibiotic efflux pumps corroborate current knowledge of Pseuomonas' pathogenicity in humans [1]. Its ability to thrive in soil environments is meanwhile evidenced by the presence of nitrile-hydrolyzing enzymes, demonstrated by its nitA gene encoding for aliphatic nitrilase. Finally, the presence of dctB and dctD (C4-dicarboxylate transport sensor protein and transcriptional regulatory protein, respectively) as well as kgtP (ketoglutarate alpha) for dicarboxylate transport hint at Pseudomonas' potential virulence in plants, as mutations in the Dct system can negatively impact bacterial growth, virulence in plants by hindering the acquisition of carbon from plants. [2]
